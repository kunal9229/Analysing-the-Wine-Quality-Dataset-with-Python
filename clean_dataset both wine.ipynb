### Data Visualization and Analysis

```python
def plot_missing_heatmap(df, title):
    plt.figure(figsize=(8, 6))
    sns.heatmap(df.isnull(), yticklabels=False, annot=True)
    plt.title(title)
    plt.show()
    
file_paths
```

```python
def checkmissing(file_paths):
    count_df = 0
    missing_value = ["na","N/a",np.nan]
    for file_path in file_paths:
        count_df+=1
        try:
            print(f"count_dataframes = {count_df}")
            df = pd.read_csv(file_path, na_values=missing_value)
            title = f"Missing Values Heatmap for {file_path}"
            plot_missing_heatmap(df, title)
        except pd.errors.ParserError:
            print(f"Error reading file: {file_path}")
```

### Relationship between the different features and wine quality

```python
def fixedAcidity_quality(dataframes):
    df_name()
    count_df = 0
    for df in dataframes:
        count_df += 1
        
        plt.figure()
        
        plt.title('Relation of Fixed Acidity with Wine')
        plt.xlabel('Quality')
        plt.ylabel('Fixed Acidity')
        plt.scatter(df['quality'], df['fixed acidity'], color='blue', label=f'DataFrame {count_df}')
        
        plt.legend()
        
        try:
            print(f"count_dataframes = {count_df}")
            plt.show()
            
        except pd.errors.ParserError:
            print(f"Error reading file: {file_path}")
        
        if count_df == 2:
            break

fixedAcidity_quality(dataframes)

```

```python
def plot_bar(dataframes, x_column, y_column):
    df_name()
    count_df = 0
    for df in dataframes:
        plt.figure()  # Create a new figure for each DataFrame
        count_df += 1
        plt.figure(figsize=(9, 6))  # Set the figure size
        plt.xlabel(x_column)
        plt.ylabel(y_column)
        plt.title(f'Bar Plot: Relationship between {y_column} and wine {x_column} (DataFrame {count_df})')
        plt.scatter(df[x_column].values, df[y_column].values, c=df['quality'].values, cmap=plt.cm.cubehelix)
        cbar = plt.colorbar()
        cbar.set_label("Quality")
        
        try:
            print(f"count_dataframes = {count_df}")
            plt.show()
            
        except pd.errors.ParserError:
            print(f"Error reading file: {file_path}")
            
        if count_df ==2:
            break;

plot_bar(dataframes, 'volatile acidity', 'alcohol')

```

```python
def alcohol_quality(dataframes):
    df_name()
    
    count_df = 0
    for df in dataframes:
        count_df += 1
        
        plt.figure()
        
        plt.title('relation of alcohol with wine')
        plt.xlabel('quality')
        plt.ylabel('alcohol')
        plt.bar(df['quality'], df['alcohol'], color = 'orange')

        plt.legend()
        
        try:
            print(f"count_dataframes = {count_df}")
            plt.show()
            
        except pd.errors.ParserError:
            print(f"Error reading file: {file_path}")
        
        if count_df == 2:
            break

alcohol_quality(dataframes)



# checking the variation of fixed acidity in the different qualities of wine
```

As we can see, the above two items do not have a very strong relationship to the dependent variable, thus we must display a correlation plot to determine which items are more associated to the dependent variable and which items are less related to the dependent variables.

```python
def heatmap_quality(dataframes):
    df_name()
    count_df = 0
    for df in dataframes:
        count_df += 1
        
        plt.figure()
        f, ax = plt.subplots(figsize=(10, 8))
        corr = df.corr()
        sns.heatmap(corr, mask=np.zeros_like(corr, dtype=np.bool_), cmap=sns.diverging_palette(220, 10, as_cmap=True),
            square=True, ax=ax)

        plt.legend()
        
        try:
            print(f"count_dataframes = {count_df}")
            plt.show()
            
        except pd.errors.ParserError:
            print(f"Error reading file: {file_path}")
        
        if count_df == 2:
            break

heatmap_quality(dataframes)

```

#### We can easily identify which items have a strong relationship to one another and which items are related to one another on a weekly basis from the correlation plot shown above for the dataset used to estimate the quality of the wine. For instance,

### The strongly correlated items are :
1.fixed acidity and citric acid. 2.free sulphur dioxide and total sulphor dioxide. 3.fixed acidity and density. 4. alcohol and quality.

Therefore, it is evident from the information above that alcohol is the most crucial factor in determining the quality of wine.

### The weekly correlated items are :
1.citric acid and volatile acidity. 2.fixed acidity and ph. 3.density and alcohol.

These relationships don't rely on one another at all.

```python
def wine_rating(dataframes):
    df_name()
    count_df = 0
    for df in dataframes:
        count_df += 1
        
        plt.figure()
        plt.hist(df['quality'], range=(1, 10))
        plt.xlabel('Ratings of wines')
        plt.ylabel('Amount')
        plt.title('Distribution of wine ratings')

        plt.legend()
        
        try:
            print(f"count_dataframes = {count_df}")
            plt.show()
            
        except pd.errors.ParserError:
            print(f"Error reading file: {file_path}")
        
        if count_df == 2:
            break

wine_rating(dataframes)

```

```python
def pairplots(dataframes):
    df_name()
    count_df = 0
    for df in dataframes:
        count_df += 1
        
        plt.figure()
        sns.pairplot(df)

        plt.legend()
        
        try:
            print(f"count_dataframes = {count_df}")
            plt.show()
            
        except pd.errors.ParserError:
            print(f"Error reading file: {file_path}")
        
        if count_df == 2:
            break

pairplots(dataframes)
```
